## Séléction des clés primaires pour les différentes tables

```python
# Clé primaire de la table "population" 
CREATE TABLE population (
    pays TEXT,
    code_pays INT,
    annee INT,
    population INT,
    PRIMARY KEY(code_pays, annee));
```

```python
# Clé primaire de la table "dispo_alim" 
CREATE TABLE dispo_alim (
    pays TEXT,
    code_pays INT,
    annee INT,
    produit TEXT,
    code_produit INT,
    origin TEXT,
    dispo_alim_tonnes FLOAT,
    dispo_alim_kcal_p_j FLOAT,
    dispo_prot FLOAT,
    dispo_mat_gr FLOAT,
    PRIMARY KEY(code_pays, code_produit, annee));
```

```python
# Clé primaire de la table "equilibre_prod" 
CREATE TABLE equilibre_prod (
    pays TEXT,
    code_pays INT,
    annee INT,
    produit TEXT,
    code_produit INT,
    dispo_int FLOAT,
    alim_ani FLOAT,
    semences FLOAT,
    pertes FLOAT,
    transformes FLOAT,
    nourriture FLOAT,
    autres_utilisations FLOAT,
    PRIMARY KEY (code_pays, code_produit, annee));
```

```python
# Clé primaire de la table "sous_nutrition" 
CREATE TABLE sous_nutrition (
    pays TEXT,
    code_pays INT,
    annee TEXT,
    nb_personnes FLOAT,
    PRIMARY KEY (code_pays, annee));
```
